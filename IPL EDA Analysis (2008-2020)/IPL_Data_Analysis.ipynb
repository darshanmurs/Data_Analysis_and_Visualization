```python
# Importing all required Libraries for Analysis and Visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading Data set for Analysis
ipl_matches = pd.read_csv("IPL Matches 2008-2020.csv")
ipl_matches
```

```python
# To get information about all the columns in the dataset
ipl_matches.info()
```

```python
# To obtain stats related to the dataset
ipl_matches.describe(include='all')
```

```python
# To obtain the columns or attributes having null values
ipl_matches.isnull().sum()
```

```python
#Identifying all the unique no of teams that played in IPL from 2008-2020
teams = list(set(ipl_matches['team1'].tolist()))
teams
```

```python
# Understanding the total number of team count participated in IPL from 2008-2020
print(len(teams))
```

```python
# Total no of Matches Played by each Individual team
x = ipl_matches['team1'].value_counts()
y = ipl_matches['team2'].value_counts()
z = x+y
z
```

```python
#Plotting Horizontal Bar Graph for matches played by each team
z.plot( kind = 'barh')
plt.ylabel('Teams')
plt.xlabel('Count')
```

```python
#Calculating total number of wins by each individual teams
winner = pd.DataFrame({"Win":ipl_matches['winner']}).value_counts()
winner
```

```python
# Plot representing most number of wins by each individual team
sns.histplot(x='winner', data=ipl_matches, hue='winner', legend=False)
plt.xticks(rotation=90)
plt.xlabel('Teams')
```

```python
# Number of matches held at different venues
place = pd.DataFrame({"Venue":ipl_matches['venue']}).value_counts()
place
sns.histplot(x='venue', data=ipl_matches, hue='venue', legend=False)
plt.xticks(rotation=90)
plt.xlabel('Match_venues')
```

```python
# Top 10 Most Man of the Match winners in IPL
mom = ipl_matches['player_of_match'].value_counts().head(10)
mom
sns.barplot(x=mom.index, y=mom.values, data=ipl_matches)
plt.xticks(rotation='vertical')
plt.xlabel("Player Names")
plt.ylabel("MoM count")
```

```python
u1 = ipl_matches['umpire1'].value_counts()
u2 = ipl_matches['umpire2'].value_counts()
U = (u1+u2)
U=U.sort_values().head(10)
U.plot(kind='barh')
plt.xlabel('Count')
plt.ylabel('Umpire_Names')
```

```python
# Calculating wins by each team every year
team_wins_per_season = ipl_matches.groupby('year')['winner'].value_counts()
team_wins_per_season
```

```python
# Taking out teams with max wins in each year for plotting graph

year = 2008
team_max_win_year = pd.DataFrame(columns=['year', 'team', 'wins'])
for i in team_wins_per_season.iteritems():
    if i[0][0]==year:
        print(i)
        team_max_wins = pd.DataFrame({
            'year': [i[0][0]],
            'team': [i[0][1]],
            'wins': [i[1]]
        })
        team_max_win_year = team_max_win_year.append(team_max_wins)
        year += 1

team_max_win_year
```

```python
sns.barplot(x ='wins', y='team', hue='year', data=team_max_win_year)
```

#Delivery DataSet Analysis

```python
#Loading Dataset
ipl_delivery_data=pd.read_csv('IPL Ball-by-Ball 2008-2020.csv')
ipl_delivery_data
```

```python
#Viewing top level data
ipl_delivery_data.head()
```

```python
# Understanding information about the data file
ipl_delivery_data.info()
```

```python
# To understand number of rows and columns in data

ipl_delivery_data.shape
```

```python
#Top runs scored by batsman across the league
top_run_scorer = ipl_delivery_data.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
top_run_scorer
```

```python
# Plotting map for top run scorers in IPL
top_run_scorer.plot(kind='bar')
```

```python
#Understanding bolwers who has leaked most runs throughout IPL
bowler_runs = ipl_delivery_data.groupby('bowler')['total_runs'].sum().sort_values(ascending=False).head(10)
bowler_runs
```

```python
# Plotting bar graph to understand runs given by individual bowlers vs each team
runs_leaked=ipl_delivery_data['bowler']=='PP Chawla'
ipl_delivery_data[runs_leaked].groupby('batting_team')['total_runs'].sum().plot(kind='bar')
```

```python
# Understanding runs scored by each teams in different overs
runs_by_over = ipl_delivery_data[runs_leaked]
runs_by_over_pivot = runs_by_over.pivot_table(values='batsman_runs', index='batting_team', columns='over', aggfunc='count')
runs_by_over_pivot
```

```python
#Plotting runs scored by each team in different overs with Heatmap
sns.heatmap(runs_by_over_pivot, cmap='Blues', annot=True)
```

```python
# Finding type of dismissal of batsman in the league
sns.histplot(x='dismissal_kind', data=ipl_delivery_data, hue='dismissal_kind')
plt.xticks(rotation='vertical')
```
